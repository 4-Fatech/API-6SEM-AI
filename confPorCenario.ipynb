```python
import os
from ultralytics import YOLO
```

# Função para calcular a confiabilidade média

```python
def calcular_acuracia(results):
    total_confidence = 0
    total_detections = 0
    for result in results:
        for box in result.boxes:
            total_confidence += box.conf
            total_detections += 1
    return total_confidence / total_detections if total_detections > 0 else 0
```

# Função para processar todos os vídeos na pasta especificada

```python
def processar_videos(pasta_videos, model):
    videos = [os.path.join(pasta_videos, video) for video in os.listdir(pasta_videos) if video.endswith('.mp4')]
    acuracias = []
    for video in videos:
        results = model.predict(source=video, stream=True)
        acuracia = calcular_acuracia(results)
        acuracias.append((video, acuracia))
    return acuracias
```

# Função para formatar e imprimir a acurácia de cada vídeo

```python
def formatar_acuracia(acuracias):
    for video, acuracia in acuracias:
        print(f'Acurácia do vídeo {video} é de {acuracia}%')
```

```python
# Caminho para a pasta que contém os vídeos
pasta_videos = './videos'
# Instância do modelo
videoInstance = YOLO("./runs/detect/train2/weights/best.pt")
# Processar vídeos e calcular acurácias
acuracias = processar_videos(pasta_videos, videoInstance)
# Imprimir resultados formatados
```

```python
formatar_acuracia(acuracias)
```
